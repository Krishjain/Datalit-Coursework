```
import tweepy
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```
#My Twitter API Authentication Variables
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''
```

```
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)

tweets = api.search('ronaldo', count=200)

data = pd.DataFrame(data=[tweet.text for tweet in tweets], columns=['Tweets'])

display(data.head(10))

print(tweets[0].created_at)
```

```
import nltk
nltk.download('vader_lexicon')
!pip install seaborn

import seaborn as sns
```

```
sid = SentimentIntensityAnalyzer()

def sentiment_analyzer_scores(text):
    score = sid.polarity_scores(text)
    lb = score['compound']
    if lb >= 0.05:
        return 1
    elif (lb > -0.05) and (lb < 0.05):
        return 0
    else:
        return -1

listy = []
fin = []

for index, row in data.iterrows():
  count = sentiment_analyzer_scores(row["Tweets"])
  fin.append(count)
  ss = sid.polarity_scores(row["Tweets"])
  listy.append(ss)
  
se = pd.Series(listy)
se1 = pd.Series(fin)
data['polarity'] = se.values
data['Sentiment'] = se1.values

display(data.head(100))
```

```
ax = sns.distplot(fin,hist=False, rug=True)
ax.set(xlabel='Negative                Neutral                 Positive',ylabel='#Tweets',title="Public Opinion on Ronaldo")
   
```
